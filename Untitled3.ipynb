```python
df=pd.read_csv('/content/drive/MyDrive/Data/Fakenews_train.csv')
```

```python
## Drop Nan values
df=df.dropna()
```

```python
## Get the Independent features
X=df.drop('label',axis=1)
```

```python
## Get the dependent features
y=df['label']
```

```python
import tensorflow as tf
```

```python

from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
#vocabulary size
voc_size=5000
```

```python
messages=X.copy()
```

```python
messages['title'][1]
```

```python
messages
```

```python
messages.reset_index(inplace=True)
messages
```

```python
import nltk
nltk.download('stopwords')
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
corpus=[]
for i in range(0,len(messages)):
  print(i)
  review=re.sub('[^a-zA-Z]',' ',messages['title'][i]) #Regular expression. substitute
  review=review.lower()
  review=review.split()
  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
one_hot_repr=[one_hot(words,voc_size) for words in corpus]
one_hot_repr
```

```python
# Embedding Representation
```

```python
#All the sentences length is different so to make it all the same we use pad sequences
```

```python
sent_length=20
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
#creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100)) #1 LSTM layer with 100 neurons
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python

import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final
```

```python
X_final.shape,y_final.shape
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python

### Finally Training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
#Adding Dropout layer
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.5))
model.add(LSTM(100))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
y_pred=model.predict_classes(X_test)
```

```python

from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
